```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python

loc = sys.argv[1]
loc = "IZH.csv"
df = pd.read_csv(loc, index_col=0)
```

```python
list(df.script_name.unique())
```

```python

sizes = df["size"].unique()
sizes.sort()

groups = df.groupby(["size"])

all_df = []
all_index = []
for size in sizes:
    group = groups.get_group(size)
    mean_time = group.groupby(["script_name"])["time"].mean()
    std_time = group.groupby(["script_name"])["time"].std()
    ratio_time = mean_time.max() / mean_time
    max_index = mean_time.idxmin()
    mean_time = mean_time.round(3)
    std_time = std_time.round(3)
    ratio_time = ratio_time.round(3)
    text_time = mean_time.astype("str") + " " + "(" + ratio_time.astype("str") + "×)"
    ratio_time = "(" + ratio_time.astype("str") + "×)"
    text_time[max_index] = "textbf" + text_time[max_index] + ";.;"
    ratio_time[max_index] = "textbf" + ratio_time[max_index] + ";.;"
    new_df = pd.DataFrame({"time": text_time})
    all_df.append(new_df.T)
    all_index.append(size)
    # new_df.to_csv(f"{loc[:-4]}/{size}.csv")

all_df = pd.concat(all_df, keys=all_index)
```

```python
all_df = all_df.droplevel(1)
all_df.columns.name=''

#set the order
all_df = all_df[['annarchy_izh_cpu.py', 'annarchy_izh_cuda.py', 'brian_izh.py', 'brian_izh_cpp.py', 'brian_izh_cuda.py', 'pynn_nest_izh.py', 'pymonnto_izh.py', 'pymonntorch_izh_cpu.py', 'pymonntorch_izh_cuda.py']]


all_df
```

```python
latex_out = all_df.to_latex(float_format="{:.3f}".format)
latex_out = latex_out.replace(";.;", "}").replace("textbf","\\textbf{").replace("emptyindex", "\\multicolumn{1}{|l|}{}")
latex_out = latex_out.replace("SecondNaN", "{ }").replace("NaN","—").replace("beforesize","\\hline \n \\multicolumn{1}{|l|}{").replace("\\\\","\\\\ \n \\hline")
print(latex_out)
```
